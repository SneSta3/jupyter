# Dodavanje novih kolona

```python
filmovi['Type'] = 'movie/tv show'
filmovi.head()
```

```python
filmovi['new_run_time'] = filmovi['RunTime'] * 10
filmovi.head()
```

```python
filmovi['NEW_GENRE'] = filmovi['GENRE'].str.strip()
filmovi.head()
```

# Promena postojece kolone

```python
filmovi['ONE-LINE'] = filmovi['ONE-LINE'].str.strip()
```

```python
filmovi['STARS'] = filmovi['STARS'].str.strip()
```

```python
filmovi.head()
```

```python
filmovi['STARS'] = filmovi['STARS'].str.replace('\n',' ')
```

```python
filmovi.head()
```

```python
filmovi['GENRE'] = filmovi['GENRE'].str.replace('\n',' ')
```

```python
filmovi.head()
```
