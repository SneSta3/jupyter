## Vizualizacija podataka

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

## Line chart

```python
x = [1,2,3]
y = [2,4,5]
plt.plot(x,y)
plt.show
```

```python
x = [1,2,3]
y = [2,4,5]
plt.xlabel('Nasa X osa')
plt.ylabel('Nasa Y osa')
plt.plot(x,y)
plt.show
```

```python
x1 = [1,2,3]
x2 = [1,2,3]
y1 = [2,4,5]
y2 = [5,3,7]
plt.xlabel('Nasa X osa')
plt.ylabel('Nasa Y osa')
plt.title('Nas prvi grafik')
plt.plot(x1,y1, label = 'Prva linija', color = 'red')
plt.plot(x2,y2, label = 'Druga linija', color = 'green')
plt.legend()
plt.show
```

## Bar Chart

```python
x = ['A','B','C','D']
y = [2,5,1,4]
plt.xlabel = ('Prva linija')
plt.ylabel = ('Druga linija')
plt.title('Bar chart')
plt.bar(x,y)
plt.show()
```

## Pie chart

```python
delovi = [4,15,10,5]
imena_delova = ['hodnik', 'dnevna soba', 'spavaca soba', 'kupatilo']
plt.pie(delovi, labels = imena_delova)
plt.show
```

```python
delovi = [4,15,10,5]
colors = ['red', 'blue', 'green', 'yellow']
imena_delova = ['hodnik', 'dnevna soba', 'spavaca soba', 'kupatilo']
plt.pie(delovi, labels = imena_delova,colors = colors)
plt.title('Pie chart')
plt.show
```

## Scatterplot

```python
x = [1,2,4,6,3,7,8,9]
y = [3,2,6,1,8,5,2,3]
plt.title('Scatterplot')
plt.scatter(x,y)
plt.show()
```
